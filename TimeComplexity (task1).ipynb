```python
# Function to create array with length = n
def n_array(n):
    return np.random.rand(n)
```

```python
# Constant function
def const(array):
    const = 1
#    new_array = []
#    for i in array:
#        new_array.append(const)
    return np.full(len(array), const)
```

```python
# The sum of elements
def sum_elem(array):
    sum_ = 0
    new_array = []
    for i in array:
        sum_ += i
    for i in range(len(array)):
        new_array.append(sum_)
    return new_array
```

```python
# The product of elements
def prod_elem(array):
    prod = 1
    new_array = []
    for i in array:
        prod *= i
    for i in range(len(array)):
        new_array.append(prod)
    return new_array
```

```python
# Naive_Polynomial
def Naive_Poly(array):
    y = 0
    for i in range(len(array)):
        p = 1
        for j in range(i):
            p = p * 1.5
        y += array[i] * p
    return y
```

```python
def Horner_Poly(array):
    y = array[0]
    for i in range(1, len(array)):
        y = y * 1.5 + array[i]
    return y
```

```python
def Bubble_sort(array):
    n = len(array)
    for i in range(n - 1):
        for j in range(0, n - i - 1):
            if array[j] > array[j + 1]:
                array[j], array[j + 1] = array[j + 1], array[j]
```

```python
def partition(array, low, high):
    
    i = (low - 1)
    temp = array[high]
    for j in range(low, high):
        if array[j] <= temp:
            i += 1
            array[i], array[j] = array[j], array[i]
    
    array[i + 1], array[high] = array[high], array[i + 1]
    return i + 1 

def Quick_sort(array, low, high):
    if len(array) == 1:
        return array
    if low < high:
        part = partition(array, low, high)
        Quick_sort(array, low, part - 1)
        Quick_sort(array, part + 1, high)
```

```python
def Tim_sort(array):
    array.sort()
```

```python
def matmul(A, B):
    zip_b = zip(*B)
    zip_b = list(zip_b)
    return [[sum(ele_a*ele_b for ele_a, ele_b in zip(row_a, col_b)) 
             for col_b in zip_b] for row_a in A]
```

```python
time_tim = []
```

```python
MINIMUM= 32
  
def find_minrun(n): 
  
    r = 0
    while n >= MINIMUM: 
        r |= n & 1
        n >>= 1
    return n + r 
  
def insertion_sort(array, left, right): 
    for i in range(left+1,right+1):
        element = array[i]
        j = i-1
        while element<array[j] and j>=left :
            array[j+1] = array[j]
            j -= 1
        array[j+1] = element
    return array
              
def merge(array, l, m, r): 
  
    array_length1= m - l + 1
    array_length2 = r - m 
    left = []
    right = []
    for i in range(0, array_length1): 
        left.append(array[l + i]) 
    for i in range(0, array_length2): 
        right.append(array[m + 1 + i]) 
  
    i=0
    j=0
    k=l
   
    while j < array_length2 and  i < array_length1: 
        if left[i] <= right[j]: 
            array[k] = left[i] 
            i += 1
  
        else: 
            array[k] = right[j] 
            j += 1
  
        k += 1
  
    while i < array_length1: 
        array[k] = left[i] 
        k += 1
        i += 1
  
    while j < array_length2: 
        array[k] = right[j] 
        k += 1
        j += 1
  
def tim_sort(array): 
    n = len(array) 
    minrun = find_minrun(n) 
  
    for start in range(0, n, minrun): 
        end = min(start + minrun - 1, n - 1) 
        insertion_sort(array, start, end) 
   
    size = minrun 
    while size < n: 
  
        for left in range(0, n, 2 * size): 
  
            mid = min(n - 1, left + size - 1) 
            right = min((left + 2 * size - 1), (n - 1)) 
            merge(array, left, mid, right) 
  
        size = 2 * size 
```
